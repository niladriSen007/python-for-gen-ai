```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python

from nltk.stem import PorterStemmer
stemming=PorterStemmer()
for word in words:
    print(word+"---->"+stemming.stem(word))
```

```python
stemming.stem("finalized"),stemming.stem("goes"),stemming.stem("went"),stemming.stem("gone")
```

## Regex stemmer


```python
from nltk.stem import RegexpStemmer
regex_stemmer = RegexpStemmer('ing$|s$|ed$|able$', min=4)
for word in words:
    print(word+"---->"+regex_stemmer.stem(word))
    
```

## Snowball Stemmer


```python
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')
for word in words:
    print(word+"---->"+snowball_stemmer.stem(word))
```

```python
snowball_stemmer.stem("finalized"),snowball_stemmer.stem("goes"),snowball_stemmer.stem("went"),snowball_stemmer.stem("gone"),snowball_stemmer.stem("eaten"),snowball_stemmer.stem("fairly")
```
