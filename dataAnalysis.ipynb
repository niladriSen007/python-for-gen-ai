```python
import pandas as pd
df = pd.read_csv("ecommerce.csv")
df.head(5)
df.tail(5)
df.describe()
```

```python
df.dtypes
```

```python
## Handling missing values
df.isnull()
## df.isnull().sum()
```

```python
""" Fill the missisng values by 0 """
newData = df.fillna(0)
newData.isnull().sum()
```

```python
## Fill the missing values by the mean of the column
newData['Price_fillNa'] = df['Price'].fillna(df['Price'].mean())
newData
```

```python
df.dtypes
```

```python
## Rename columns
newData = newData.rename(columns={'Price_fillNa':'Price without NA'})
newData
```

```python
## change the data types of columns
newData['Price'] = newData['Price'].astype('int')
newData.dtypes
```

```python
newData['Increased_price'] = newData['Price'].apply(lambda x: x*10)
newData
```

```python
## aggregate data
groupedData = newData.groupby('Payment_Method')['Price'].agg(['mean', 'sum','count'])
groupedData
groupedData = groupedData.iloc[1:]
groupedData
```

```python
##merge data and join data
data1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'key': ['K0', 'K1', 'K2', 'K3']})
data2 = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3'],
                      'key': ['K0', 'K1', 'K2', 'K3']})
pd.merge(data1, data2, on='key',how='inner')
```
