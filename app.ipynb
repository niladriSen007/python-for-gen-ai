```python
tuple(map(lambda x: x**2, [1,2,3,4,5]))
tuple(map(lambda name: name.capitalize(), ['ab','bb','cn','dm','ed']))
tuple(map(lambda obj: obj['name'].upper(), [{'name':'ab'},{'name':'bb'},{'name':'cn'},{'name':'dm'},{'name':'ed'}]))
```

```python
num1 = [1,2,3,4,5]
num2 = [6,7,8,9,10]
num = set(map(lambda x,y: x+y, num1,num2))
print(num)
```

```python
str_to_int_1 = map(lambda x: int(x), ['1','2','3','4','5'])
str_to_int_2 = map(int, ['1','2','3','4','5'])
print(tuple(str_to_int_1))
print(tuple(str_to_int_2))
```
